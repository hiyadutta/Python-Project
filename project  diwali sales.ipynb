## Importing libraries

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing the Diwali Sales Data
df= pd.read_csv("1. Diwali Sales Data_python.csv", encoding="unicode_escape") 
```

```python
df
```

## Data Cleaning

```python
df.shape
```

```python
df.head(10)
```

```python
# Checking for null values 
df.info()
```

```python
df.drop(["Status","unnamed1"], axis=1, inplace=True)   # delete the Unnamed1 Coloumn as it is irrelevent in the dataset 
```

```python
df.info()
```

```python
print("no of duplicated data present in the dataset: ", df.duplicated().sum()) ## Total duplicate data in the dataset
```

```python
df = df.drop_duplicates()  ## Droping the duplicate values
```

```python
print("no of duplicated data present in the dataset: ", df.duplicated().sum())   ## After dropping the duplicate values 
```

```python
pd.isnull(df).sum()/len(df)* 100   # Checking for percentage of null values present in a coloumn
```

```python
df.shape
```

```python
sns.boxplot(x='Amount',data=df)
plt.show()
# In the amount coloum outlier is present so we can use median to impute the null values
```

```python
df['Amount'] = df['Amount'].fillna(df['Amount'].median())  
## Impute the null values of the Amount coloum with the median values
```

```python
pd.isnull(df).sum()/len(df)* 100   ## After imputing null values 
```

### Gender count

```python
ax=sns.countplot(x="Gender", data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs we can see that most of the buyers are females 

### Gender wise Amount

```python
sales_gender=df.groupby(['Gender'],as_index=False).Amount.sum().sort_values(by='Amount',ascending=True)
```

```python
sales_gender
grapy=sns.barplot(x='Gender',y='Amount',data=sales_gender)
for bars in grapy.containers:
    grapy.bar_label(bars)
```

From above graphs we can see that purchasing power of females are greater than men

```python
df.columns
```

### Age group wise Gender Count

```python
ax=sns.countplot(data=df, x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs we can see that most of the buyers are of age group between 26-35 years

### Order Amount by Age group

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum()
```

```python
sales_age
```

```python
sns.set(rc={'figure.figsize':(20,10)})
Graph_salesAge = sns.barplot(y= 'Amount', x= 'Age Group', data =sales_age)
for i in Graph_salesAge.containers:
    Graph_salesAge.bar_label(i)
```

### Orders per State

```python
state_orders=df.groupby('State',as_index=False).Orders.sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
state_orders
```

```python
sns.set(rc={'figure.figsize':(20,10)})
sales_G = sns.barplot(data=state_orders,x='State',y='Orders')
sales_G

for i in sales_G.containers:
    sales_G.bar_label(i)
```

```python
##State wise total amount


State_amount = df.groupby('State',as_index=False).Amount.sum().sort_values(by = 'Amount',ascending= False).head(10)
State_amount
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sales_am_graph= sns.barplot(data=State_amount,x= 'State',y='Amount')

for i in sales_am_graph.containers:
    sales_am_graph.bar_label(i)
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

### Purchasing power based on gender and marital status

```python
amount_gen = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
amount_gen
```

```python
amt_gen = sns.barplot(data= amount_gen, x= 'Gender', y= 'Amount', hue =   'Marital_Status' )
amt_gen


for i in amt_gen.containers:
    amt_gen.bar_label(i)
```

Single females have a high purchasing power

## Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

## Product Category

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

## top 10 most sold products 

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
